```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_data.csv")
```

```python
df
```

```python
print(df.Country.unique())
```

#### let's fix the column name (it's too long):

```python
df.rename(columns={'Life expectancy at birth (years)': 'Life_exp'}, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

# Life expectancy

```python
plt.figure(figsize=(9,7))
sns.lineplot(data=df,x="Year",y="Life_exp")
plt.xlabel("Year")
plt.ylabel("Life expectancy at birth (years)")
plt.title("life expectancy over time in the six nation")
```

```python
plt.figure(figsize=(9,7))
sns.relplot(data=df,x="Year", y="Life_exp", hue="Country", kind="line")
plt.ylabel("Life expectancy at birth (years)")
plt.title("life expectancy over time in the six nation")
```

```python
g = sns.FacetGrid(df,col="Country",col_wrap=3,sharey = False, sharex = False,hue="Country",height=4.5)
g.map(sns.lineplot,"Year","Life_exp")
g.add_legend()
g.set_xlabels("Life expectancy at birth (years)")
g.set_ylabels("GDP (trillion dollars)")
#g.set_title("life expectancy over time")
g.fig.subplots_adjust(top=0.9) # adjust the Figure in rp
g.fig.suptitle('Life expectancy over time')
```

```python
plt.figure(figsize=(9,7))
ax = plt.subplot()
sns.boxplot(y="Life_exp",x="Country",data=df)
plt.ylabel("Life expectancy at birth (years)")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(9,7))
ax = plt.subplot()
sns.boxplot(y="Life_exp",x="Country",data=df)
plt.ylabel("Life expectancy at birth (years)")
plt.xticks(rotation=90)
sns.stripplot(y="Life_exp", x="Country", data=df,
              size=3, color=".3", linewidth=1)
```

```python
plt.figure(figsize=(9,7))
ax = plt.subplot()
sns.violinplot(y="Life_exp",x="Country",data=df)
plt.ylabel("Life expectancy at birth (years)")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(9,7))
ax = plt.subplot()
sns.violinplot(y="Life_exp",x="Country",data=df)
plt.ylabel("Life expectancy at birth (years)")
plt.xticks(rotation=90)
sns.stripplot(y="Life_exp", x="Country", data=df,
              size=3, color=".8", linewidth=1)
```
